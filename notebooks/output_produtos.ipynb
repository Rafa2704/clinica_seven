```python
import pandas as pd

# Caminho do arquivo CSV
caminho_arquivo = r"C:\Users\rafae\Desktop\clinica_sevem\Data_Lake\products_raw.csv"

# Leitura do arquivo CSV
df = pd.read_csv(caminho_arquivo)

df
```

```python
df_wide = df.pivot_table(
    index='product_id',
    columns='var_name',
    values='var_value',
    aggfunc='first'  # ou 'max', 'min', 'sum', dependendo do contexto
).reset_index()
```

```python
df_wide
```

```python
print(df_wide.info())
```

```python
colunas_data = ['created_at']
for col in colunas_data:
    df_wide[col] = pd.to_datetime(df_wide[col], errors='coerce', dayfirst=True)
```

```python
# 2. Limpar coluna "total" e converter para float
df_wide['price'].replace('R\$', '', regex=True, inplace=True)       # Remove "R$"
df_wide['price'].replace('\.', '', regex=True, inplace=True)        # Remove ponto (milhar)
df_wide['price'].replace(',', '.', regex=True, inplace=True)        # Troca vírgula decimal por ponto
df_wide['price'] = df_wide['price'].astype(float)   
```

```python
df_wide
```

```python
print(df_wide.columns)
```

```python
# Remove o nome do eixo das colunas (var_name)
df_wide.columns.name = None

# Copia para df_final
df_final = df_wide.copy()
```

```python
df_final
```

```python
import re
def limpar_nome(nome):
    if pd.isnull(nome):
        return nome
    # Remove tudo que não for letra, número ou espaço
    nome_limpo = re.sub(r'[^\w\s]', '', nome)
    # Substitui underscores por espaço
    nome_limpo = nome_limpo.replace('_', ' ')
    # Capitaliza
    return nome_limpo.title()
```

```python
df_final['name'] = df_final['name'].apply(limpar_nome)
```
