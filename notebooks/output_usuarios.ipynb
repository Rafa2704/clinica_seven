```python
import pandas as pd

# Caminho do arquivo CSV
caminho_arquivo = r"C:\Users\rafae\Desktop\clinica_sevem\Data_Lake\user_raw.csv"

# Leitura do arquivo CSV
df = pd.read_csv(caminho_arquivo)

df
```

```python
df_wide = df.pivot_table(
    index='user_id',
    columns='var_name',
    values='var_value',
    aggfunc='first'  # ou 'max', 'min', 'sum', dependendo do contexto
).reset_index()
```

```python
df_wide
```

```python
df_wide.columns.name = None
```

```python
df_wide
```

```python
import re
def limpar_nome(nome):
    if pd.isnull(nome):
        return nome
    # Remove tudo que não for letra, número ou espaço
    nome_limpo = re.sub(r'[^\w\s]', '', nome)
    # Substitui underscores por espaço
    nome_limpo = nome_limpo.replace('_', ' ')
    # Capitaliza
    return nome_limpo.title()
```

```python
df_wide['name'] = df_wide['name'].apply(limpar_nome)
```

```python
df_wide
```

```python
df_wide = df_wide[df_wide['e-mail'].str.contains(r'@exemplo\.com$', na=False)]
```

```python
df_wide
```

```python
df_wide['entry_time'] = df_wide['entry_time'].str.strip()

# Se não tiver segundos, adiciona ":00"
df_wide['entry_time'] = df_wide['entry_time'].apply(
    lambda x: x if len(x.split(':')) == 3 else x + ':00'
)
```

```python
df_wide
```

```python
df_wide['entry_time'] = df_wide['entry_time'].str.replace(r'\s?(AM|PM)', '', regex=True)
```

```python
df_wide
```

```python
print(df_wide.info())
```

```python
# Converter para datetime, tratando formatos diferentes
df_wide['entry_date'] = pd.to_datetime(df_wide['entry_date'], errors='coerce', dayfirst=True)
df_wide['update_date'] = pd.to_datetime(df_wide['update_date'], errors='coerce', dayfirst=True)

# Formatar como string no padrão aaaa/mm/dd
df_wide['entry_date'] = df_wide['entry_date'].dt.strftime('%Y/%m/%d')
df_wide['update_date'] = df_wide['update_date'].dt.strftime('%Y/%m/%d')
```

```python
df_wide
```

```python
colunas_data = ['entry_date', 'update_date']
for col in colunas_data:
    df_wide[col] = pd.to_datetime(df_wide[col], errors='coerce', dayfirst=True)
```

```python
df_final = df_wide.copy()
```
